```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
font = {'size'   : 12}

plt.rc('font', **font)
```

```python
us_df = pd.read_csv('time_series_covid_19_confirmed_US.csv')
us_df.info()
```

```python
# us_df.head()
```

```python
# us_df.Province_State.unique()
```

## For Texas

```python
texas_df = us_df[us_df['Province_State']=='Texas']
texas_df.info()
```

```python
# texas_df
```

```python
texas_df = texas_df.drop(['UID', 'Province_State', 'Country_Region'], axis=1)
# texas_df
```

```python
tx_data = texas_df.sum(axis = 0, skipna = True) 
# tx_data
```

```python
final_data = pd.DataFrame({'Date': texas_df.columns, 'Confirmed': tx_data})
# final_data
```

```python
## Find the increase amount in each day
final_data['Increase'] =  final_data['Confirmed'] - final_data['Confirmed'].shift(1)
# final_data.head()
```

```python
final_data['Moving_Average'] = final_data.Increase.rolling(window=7).mean()
# final_data
```

```python
texas_data = final_data
texas_data.info()
```

```python
plt.figure(figsize=(20,10))
plt.bar(final_data['Date'], final_data['Increase'], color="dodgerblue", width=0.25, label = 'Confirmed Cases')
plt.plot(final_data['Date'], final_data['Moving_Average'], color="red", linewidth=5, label = '7-days moving average')
plt.xlabel('', fontsize=18)
plt.ylabel('# of Confirmed Cases', fontsize=18)
plt.title('TEXAS')
plt.xlim('3/11/20', '5/20/20')
plt.xticks(rotation=70)
plt.axvspan('3/23/20', '4/30/20', alpha=0.1, color='red')
plt.axvspan('4/30/20', '5/20/20', alpha=0.1, color='green')
plt.legend(loc='upper left', fancybox=True, framealpha=1, shadow=True, borderpad=1)
# plt.savefig('plots/Texas_reopen.png')
plt.show()
```
